```python
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
style.use("fivethirtyeight")
x,y = make_blobs(n_samples = 1000, centers=3, cluster_std = 1, n_features=2)
plt.scatter(x[:,0], x[:,1], s=4, color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
plt.clf()
```

```python
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt 
from matplotlib import style
 
X, y = make_moons(n_samples = 10000, noise = 0.1)
plt.scatter(X[:, 0], X[:, 1], s = 1, color ='g')
plt.xlabel("X")
plt.ylabel("Y")
 
plt.show()
plt.clf()
```
