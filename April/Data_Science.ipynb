```python
#y=mx+c
# m is slope
# c is intercept
# x is independent variable and y is dependent variable
```

```python
m = 2
c = 5
plt.figure(figsize=(10,5))
x = np.arange(1,10)
y = m*x + c
plt.plot(x,y, marker='o', color='g')
#Add labels and tittle
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = m*x + c')
plt.grid(True)
plt.legend()
plt.show()
```

```python
data_path = 'Salary_Data.csv'
df = pd.read_csv(data_path)
df
```

```python
x = df['YearsExperience'].values.reshape(-1,1)
print(x)
```

```python
y = df['Salary'].values
print(y)
```

```python
model = LinearRegression()
print(model)
```

```python
model.fit(x,y)
```

```python
slope = model.coef_[0]
print(slope)
```

```python
intercept = model.intercept_
print(intercept)
```

```python
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, slope * x + intercept, color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f'Equation of the regression line: y = {slope:.2f}x + {intercept:.2f}')
```

```python
#Multi Linear Regression 3*x1+2*x2+4*x3
x1 = np.linspace(100, 200, num=len(df)).reshape(-1, 1)
x2 = np.linspace(100, 200, num=len(df)).reshape(-1, 1)
x3 = np.linspace(100, 200, num=len(df)).reshape(-1, 1)
```

```python
# Calculate y values based on the equation y = 3x1 + 2x2 + 4x3 + 5
y = 3 * x1 + 2 * x2 + 4 * x3 + 5
```

```python
# Perform simple linear regression for x1, x2, x3 and y
model_x1 = LinearRegression()
model_x1.fit(x1, y)
```

```python
model_x2 = LinearRegression()
model_x2.fit(x2, y)
```

```python
model_x3 = LinearRegression()
model_x3.fit(x3, y)
```

```python
# Plot the data points
plt.scatter(x1, y, color='blue', label='x1')
plt.scatter(x2, y, color='green', label='x2')
plt.scatter(x3, y, color='orange', label='x3')

# Plot the regression lines
plt.plot(x1, model_x1.predict(x1), color='red', label='Regression Line for x1')
plt.plot(x2, model_x2.predict(x2), color='purple', label='Regression Line for x2')
plt.plot(x3, model_x3.predict(x3), color='brown', label='Regression Line for x3')

# Add labels and legend
plt.xlabel('Independent Variables')
plt.ylabel('Dependent Variable (y)')
plt.title('Multi Linear Regression')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
